```python
from tkinter import *
from tkinter import ttk
from tkinter import messagebox


class product:
    def __init__(self, sku, name, category, quantity, origin, price):
        self.sku = sku
        self.name = name
        self.category = category
        self.quantity = quantity
        self.origin = origin
        self.price = price


list_product = []


def add_info():
    list_product.append(product(entry_sku.get(), entry_name.get(), entry_category.get(), entry_quantity.get(),
                                entry_origin.get(), entry_price.get()))
    product_Table.insert(parent='', index='end', values=(
        entry_sku.get(), entry_name.get(), entry_category.get(), entry_quantity.get(), entry_origin.get(),
        entry_price.get()))


def clear_all():
    for i in product_Table.get_children():
        product_Table.delete(i)


def edit_info():
    check = entry_sku.get()
    exist = False
    for i in range(len(list_product)):
        if list_product[i].sku == check:
            list_product[i].sku = entry_sku.get()
            list_product[i].name = entry_name.get()
            list_product[i].category = entry_category.get()
            list_product[i].quantity = entry_quantity.get()
            list_product[i].origin = entry_origin.get()
            list_product[i].price = entry_price.get()
            clear_all()
            for item in list_product:
                product_Table.insert(parent='', index='end', values=(
                    item.sku, item.name, item.category, item.quantity, item.origin, item.price))
            messagebox.showinfo("Edit",
                                f"Product with SKU: {list_product[i].sku}, name: {list_product[i].name} is "
                                f"updated successfully")
            exist = True
            break
    if not exist:
        messagebox.showinfo("Edit", f"Product with SKU: {check} does not exist")


def delete_info():
    delete_sku = entry_sku.get()
    exist = False
    for i in range(len(list_product)):
        if list_product[i].sku == entry_sku.get():
            exist = True
            del list_product[i]
            clear_all()
            for item in list_product:
                product_Table.insert(parent='', index='end', values=(
                    item.sku, item.name, item.category, item.quantity, item.origin, item.price))
            messagebox.showinfo("Delete", f"Product with SKU : {delete_sku} deleted successfully")
            break
    if not exist:
        messagebox.showinfo("Delete", f"Product with SKU: {delete_sku} does not exist")


def search_info():
    search = text_Search.get()
    exist = False
    if combo_Search.get() == "Name":
        for i in list_product:
            if i.name == search:
                exist = True
                messagebox.showinfo("Search",
                                    f"Product's name: {search} exists. SKU: {i.sku}, Name: {i.name}, "
                                    f"Category: {i.category}, Quantity: {i.quantity}, Origin: {i.origin}")
                break
    elif combo_Search.get() == "SKU":
        for i in list_product:
            if i.sku == search:
                exist = True
                messagebox.showinfo("Search",
                                    f"Product's SKU: {search} exists. SKU: {i.sku}, Name: {i.name}, "
                                    f"Category: {i.category}, Quantity: {i.quantity}, Origin: {i.origin}")
                break

    if not exist:
        messagebox.showinfo("Search", f"{search} does not exist")


def reset_info():
    entry_sku.delete(0, len(entry_sku.get()))
    entry_name.delete(0, len(entry_name.get()))
    entry_category.delete(0, len(entry_category.get()))
    entry_quantity.delete(0, len(entry_quantity.get()))
    entry_origin.delete(0, len(entry_origin.get()))
    entry_price.delete(0, len(entry_price.get()))


root = Tk()
root.title("Product Management System")
root.geometry('1500x700+0+0')
title = Label(root, text="Product Management System", font=("arial", 40, "bold"), fg="white", bg="navy")
title.pack(side=TOP, fill=X)

Manager_Frame = Frame(root, bd=4, relief=RIDGE, bg="Green")
Manager_Frame.place(x=10, y=70, width=400, height=550)

m_title = Label(Manager_Frame, text="Product", font=("arial", 16, "bold"), bg="green", fg="white")
m_title.grid(row=0, columnspan=2, pady=10, padx=80)

lbl_sku = Label(Manager_Frame, text="SKU", font=("arial", 12, "bold"), bg="green", fg="white")
lbl_sku.grid(row=1, column=0, pady=10, padx=20, sticky="w")
entry_sku = Entry(Manager_Frame, font=("arial", 12), bd=5, relief=GROOVE)
entry_sku.grid(row=1, column=1, pady=10, padx=20, sticky="w")

lbl_name = Label(Manager_Frame, text="Name", font=("arial", 12, "bold"), bg="green", fg="white")
lbl_name.grid(row=2, column=0, pady=10, padx=20, sticky="w")
entry_name = Entry(Manager_Frame, font=("arial", 12), bd=5, relief=GROOVE)
entry_name.grid(row=2, column=1, pady=10, padx=20, sticky="w")

lbl_category = Label(Manager_Frame, text="Category", font=("arial", 12, "bold"), bg="green", fg="white")
lbl_category.grid(row=4, column=0, pady=10, padx=20, sticky="w")
entry_category = Entry(Manager_Frame, font=("arial", 12), bd=5, relief=GROOVE)
entry_category.grid(row=4, column=1, pady=10, padx=20, sticky="w")

lbl_quantity = Label(Manager_Frame, text="Quantity", font=("arial", 12, "bold"), bg="green", fg="white")
lbl_quantity.grid(row=5, column=0, pady=10, padx=20, sticky="w")
entry_quantity = Entry(Manager_Frame, font=("arial", 12), bd=5, relief=GROOVE)
entry_quantity.grid(row=5, column=1, pady=10, padx=20, sticky="w")

lbl_origin = Label(Manager_Frame, text="Origin", font=("arial", 12, "bold"), bg="green", fg="white")
lbl_origin.grid(row=6, column=0, pady=10, padx=20, sticky="w")
entry_origin = Entry(Manager_Frame, font=("arial", 12), bd=5, relief=GROOVE)
entry_origin.grid(row=6, column=1, pady=10, padx=20, sticky="w")

lbl_price = Label(Manager_Frame, text="Price", font=("arial", 12, "bold"), bg="green", fg="white")
lbl_price.grid(row=7, column=0, pady=10, padx=20, sticky="w")
entry_price = Entry(Manager_Frame, font=("arial", 12), bd=5, relief=GROOVE)
entry_price.grid(row=7, column=1, pady=10, padx=20, sticky="w")

# Button
Btn_Frame = Frame(root, bd=4, relief=RIDGE, bg="white")
Btn_Frame.place(x=10, y=530, width=400)

Btn_Add = Button(Btn_Frame, text="Add", width=10, command=add_info)
Btn_Add.grid(row=0, column=0, pady=10, padx=10)
Btn_Reset = Button(Btn_Frame, text="Reset", width=10, command=reset_info)
Btn_Reset.grid(row=0, column=1, pady=10, padx=10)
Btn_Edit = Button(Btn_Frame, text="Edit", width=10, command=edit_info)
Btn_Edit.grid(row=0, column=2, pady=10, padx=10)
Btn_Delete = Button(Btn_Frame, text="Delete", width=10, command=delete_info)
Btn_Delete.grid(row=0, column=3, pady=10, padx=10)

Detail_Frame = Frame(root, bd=4, relief=RIDGE, bg="Green")
Detail_Frame.place(x=420, y=70, width=850, height=550)

#Search
lbl_Search = Label(Detail_Frame, text="Search By", font=("arial", 12, "bold"), bg="green", fg="white")
lbl_Search.grid(row=0, column=0, pady=10, padx=20, sticky="w")

combo_Search = ttk.Combobox(Detail_Frame, width=10, font=("arial", 11), state="readonly")
combo_Search['values'] = ("SKU", "Name")
combo_Search.grid(row=0, column=1, ipady=1.5, ipadx=1)

text_Search = Entry(Detail_Frame, width=30, bd=5, relief=GROOVE)
text_Search.grid(row=0, column=2, pady=10, padx=20, sticky="w")

Btn_Search = Button(Detail_Frame, text="Search", width=10, pady=5, command=search_info)
Btn_Search.grid(row=0, column=3, pady=10, padx=10)

# Table
Table_Frame = Frame(Detail_Frame, bd=4, relief=RIDGE, bg="white")
Table_Frame.place(x=10, y=70, width=820, height=460)

scroll_x = Scrollbar(Table_Frame, orient=HORIZONTAL)
scroll_y = Scrollbar(Table_Frame, orient=VERTICAL)
product_Table = ttk.Treeview(Table_Frame,
                             columns=("SKU", "Name", "Category", "Quantity", "Origin", "Price"),
                             xscrollcommand=scroll_x.set, yscrollcommand=scroll_y.set)
scroll_x.pack(side=BOTTOM, fill=X)
scroll_y.pack(side=RIGHT, fill=Y)
scroll_x.config(command=product_Table.xview)
scroll_y.config(command=product_Table.yview)

product_Table.heading("SKU", text="SKU")
product_Table.heading("Name", text="Name")
product_Table.heading("Category", text="Category")
product_Table.heading("Quantity", text="Quantity")
product_Table.heading("Origin", text="Origin")
product_Table.heading("Price", text="Price")

product_Table['show'] = 'headings'
product_Table.column("SKU", width=100)
product_Table.column("Name", width=100)
product_Table.column("Category", width=100)
product_Table.column("Quantity", width=100)
product_Table.column("Origin", width=100)
product_Table.column("Price", width=200)

product_Table.pack(fill=BOTH, expand=1)

root.mainloop()
```
